```python
import pandas as pd
```

```python
df = pd.read_excel("2018 Compiled Weather Data.xlsx", index_col=0)
df.head()
```

```python
for column in df.columns:
    if column == 'Percent Output':
        break
    df[column] = (df[column] - df[column].mean()) / df[column].std()
```

```python
df.describe()
```

```python
df.drop(columns=['snow', 'snowdepth'])
```

```python
df.to_csv("2018 Compiled Weather Data.csv")
```
